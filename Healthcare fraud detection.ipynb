#### Data Pre-processing

```python
# Remove duplicates using 'ClaimID' column and keep firs occurance
train_provider_with_patient_detailsdata.drop_duplicates(subset=['ClaimID'], keep='first', inplace=True)
test_provider_with_patient_detailsdata.drop_duplicates(subset=['ClaimID'], keep='first', inplace=True)
```

```python
#### Check number of rows and columns

print('No of rows and columns in train set (After removing duplicates):',train_provider_with_patient_detailsdata.shape)
print('No of rows and columns in test (After removing duplicates):', test_provider_with_patient_detailsdata.shape)
```

```python
## Let's create a new feature Claim_Duration = Claim End Date - Claim Start Date
train_provider_with_patient_detailsdata['ClaimStartDt'] = pd.to_datetime(train_provider_with_patient_detailsdata['ClaimStartDt'], format="%Y-%m-%d")
train_provider_with_patient_detailsdata['ClaimEndDt'] = pd.to_datetime(train_provider_with_patient_detailsdata['ClaimEndDt'], format="%Y-%m-%d")

train_provider_with_patient_detailsdata['Claim_Duration'] = (train_provider_with_patient_detailsdata['ClaimEndDt'] - train_provider_with_patient_detailsdata['ClaimStartDt']).dt.days

test_provider_with_patient_detailsdata['ClaimStartDt'] = pd.to_datetime(test_provider_with_patient_detailsdata['ClaimStartDt'], format="%Y-%m-%d")
test_provider_with_patient_detailsdata['ClaimEndDt'] = pd.to_datetime(test_provider_with_patient_detailsdata['ClaimEndDt'], format="%Y-%m-%d")

test_provider_with_patient_detailsdata['Claim_Duration'] = (test_provider_with_patient_detailsdata['ClaimEndDt'] - test_provider_with_patient_detailsdata['ClaimStartDt']).dt.days
```

```python
## Let's create a new feature Treatment_Duration = DischargeDt - AdmissionDt

train_provider_with_patient_detailsdata['AdmissionDt'] = pd.to_datetime(train_provider_with_patient_detailsdata['AdmissionDt'], format="%Y-%m-%d")
train_provider_with_patient_detailsdata['DischargeDt'] = pd.to_datetime(train_provider_with_patient_detailsdata['DischargeDt'], format="%Y-%m-%d")

train_provider_with_patient_detailsdata['Treatment_Duration'] = (train_provider_with_patient_detailsdata['DischargeDt'] - train_provider_with_patient_detailsdata['AdmissionDt']).dt.days

test_provider_with_patient_detailsdata['AdmissionDt'] = pd.to_datetime(test_provider_with_patient_detailsdata['AdmissionDt'], format="%Y-%m-%d")
test_provider_with_patient_detailsdata['DischargeDt'] = pd.to_datetime(test_provider_with_patient_detailsdata['DischargeDt'], format="%Y-%m-%d")

test_provider_with_patient_detailsdata['Treatment_Duration'] = (test_provider_with_patient_detailsdata['DischargeDt'] - test_provider_with_patient_detailsdata['AdmissionDt']).dt.days

train_provider_with_patient_detailsdata['Treatment_Duration'] = train_provider_with_patient_detailsdata['Treatment_Duration'].fillna(value=0)
test_provider_with_patient_detailsdata['Treatment_Duration'] = test_provider_with_patient_detailsdata['Treatment_Duration'].fillna(value=0)
```

```python
# Let's create a new feature DiagnosisCount

diagnosis_columns = ['ClmDiagnosisCode_1', 'ClmDiagnosisCode_2', 'ClmDiagnosisCode_3', 
                     'ClmDiagnosisCode_4', 'ClmDiagnosisCode_5', 'ClmDiagnosisCode_6', 
                     'ClmDiagnosisCode_7', 'ClmDiagnosisCode_8', 'ClmDiagnosisCode_9', 
                     'ClmDiagnosisCode_10']

train_provider_with_patient_detailsdata[diagnosis_columns] = train_provider_with_patient_detailsdata[diagnosis_columns].astype(str)   # Converting all values in the diagnosis columns to a string type

# counting the number of diagnoses, excluding the 'nan' lines and empty lines
train_provider_with_patient_detailsdata['DiagnosisCount'] = train_provider_with_patient_detailsdata[diagnosis_columns].apply(
    lambda row: row.isin(['nan', '']).sum(), axis=1)

# checking the number of NaN values in the new column
print(train_provider_with_patient_detailsdata['DiagnosisCount'].isnull().sum())


test_provider_with_patient_detailsdata[diagnosis_columns] = test_provider_with_patient_detailsdata[diagnosis_columns].astype(str)   # Converting all values in the diagnosis columns to a string type

# counting the number of diagnoses, excluding the 'nan' lines and empty lines
test_provider_with_patient_detailsdata['DiagnosisCount'] = test_provider_with_patient_detailsdata[diagnosis_columns].apply(
    lambda row: row.isin(['nan', '']).sum(), axis=1)

# checking the number of NaN values in the new column
print(test_provider_with_patient_detailsdata['DiagnosisCount'].isnull().sum())
```

## Model Building

#### Creating a Basline Model

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report

# 1. Initializing the model
nb = GaussianNB()

# 2. Model training on X_train
nb.fit(X_train, y_train.values.ravel())

# 3. Evaluation of the model on X_val
y_val_pred = nb.predict(X_val)

print('\nConfusion Matrix:\n')
print(confusion_matrix(y_val, y_val_pred))
       
# 4. Metrics report on X_val
print("\nValidation Report")
print(classification_report(y_val, y_val_pred))

# 5. Evaluation of the model on X_test (final test)
y_test_pred = nb.predict(X_test)

# Creating a Data Frame with predictions
predictions_df = pd.DataFrame({'Prediction': y_test_pred})

# Saving predictions to a CSV file
predictions_df.to_csv('test_predictionsNB.csv', index=False)
```

#### Model Comparison & Selection

```python
#Let's now experiment with a few different basic models 

## Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state =32, max_depth=4)

dt.fit(X_train, y_train.values.ravel())
y_val_pred = dt.predict(X_val)
print('\nConfusion Matrix for DT:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for DT")
print(classification_report(y_val, y_val_pred))
# Evaluation of the model on X_test (final test)
y_test_pred = dt.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsDT.csv', index=False)

## Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=32, max_iter = 2000, class_weight = 'balanced')

lr.fit(X_train, y_train.values.ravel())
y_val_pred = lr.predict(X_val)
print('\nConfusion Matrix for LR:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for LR")
print(classification_report(y_val, y_val_pred))
# Evaluation of the model on X_test (final test)
y_test_pred = lr.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsLR.csv', index=False)


## KNN 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline, Pipeline

knn = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=3))

knn.fit(X_train, y_train.values.ravel())
y_val_pred = knn.predict(X_val)
print('\nConfusion Matrix for KNN:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for KNN")
print(classification_report(y_val, y_val_pred))
# Evaluation of the model on X_test (final test)
y_test_pred = knn.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsKNN.csv', index=False)

## Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train.values.ravel())

y_val_pred = rf.predict(X_val)
print('\nConfusion Matrix for RF:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for RF")
print(classification_report(y_val, y_val_pred))
# Evaluation of the model on X_test (final test)
y_test_pred = rf.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsRF.csv', index=False)
```

```python
# Import necessary libraries
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline

# Create an XGBoost model pipeline
xgb = XGBClassifier(objective='binary:logistic', 
                    eval_metric='logloss', 
                    use_label_encoder=False, 
                    n_estimators=100, 
                    max_depth=6, 
                    learning_rate=0.1, 
                    random_state=42)

# XGBoost does not require standardization/scaling, so we do not use StandardScaler
xgb_pipeline = make_pipeline(xgb)

# Train the model
xgb_pipeline.fit(X_train, y_train.values.ravel())

# Predict on the validation set
y_val_pred = xgb_pipeline.predict(X_val)
print('\nConfusion Matrix for XGBoost:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for XGBoost")
print(classification_report(y_val, y_val_pred))

# Evaluate the model on X_test (final test)
y_test_pred = xgb_pipeline.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsXGBoost.csv', index=False)
```

#### Parameter Tunning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

rf = RandomForestClassifier(random_state=42)

param_dist = {
    'n_estimators': randint(100, 200),  
    'max_depth': [10, 20, 30, 40], 
    'min_samples_split': randint(2, 6), 
    'min_samples_leaf': randint(1, 3),  
    'max_features': ['sqrt', 'log2']
}

random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=10, cv=3, n_jobs=-1, verbose=2, random_state=42)
random_search.fit(X_train, y_train.values.ravel())

print(f"Best parameters found: {random_search.best_params_}")

best_rf = random_search.best_estimator_

best_rf.fit(X_train, y_train.values.ravel())

y_val_pred = best_rf.predict(X_val)
print('\nConfusion Matrix for RF:\n')
print(confusion_matrix(y_val, y_val_pred))
print("\nValidation Report for RF")
print(classification_report(y_val, y_val_pred))

y_test_pred = best_rf.predict(X_test)
predictions_df = pd.DataFrame({'Prediction': y_test_pred})
predictions_df.to_csv('test_predictionsRF.csv', index=False)
```
